```python
def limpieza(columnas): 
    q25, q75 = np.percentile(columnas, [25 ,75])
    iqr = q75 - q25
    
    if (q25 - (1.5 * iqr)) < 0:
        IQMenor = 0.0
    else:
        IQMenor = float(q25 - (1.5*iqr))
        
    IQMayor = float(q75 + (1.5*iqr))
    
    return IQMenor, IQMayor
```

```python
IQMenor, IQMayor = limpieza(mejor.rooms)
nuevo_df= mejor
nuevo_df = mejor[(mejor.rooms > IQMenor) & (mejor.rooms < IQMayor)]
sns.boxplot(data = nuevo_df, x = "rooms")
```

```python
IQMenor, IQMayor = limpieza(nuevo_df.bedrooms)
nuevo_df = nuevo_df[( nuevo_df.bedrooms > IQMenor) & (nuevo_df.bedrooms < IQMayor)]
sns.boxplot(data = nuevo_df, x = "bedrooms")
```

```python
IQMenor, IQMayor = limpieza(nuevo_df.price)
nuevo_df = nuevo_df[( nuevo_df.price > IQMenor) & (nuevo_df.price < IQMayor)]
sns.boxplot(data = nuevo_df, x = "price")
```

```python
nuevo_df.surface_total = nuevo_df.surface_total.fillna(value=0)
IQMenor, IQMayor = limpieza(nuevo_df.surface_total)
nuevo_df = nuevo_df[( nuevo_df.surface_total > IQMenor) & (nuevo_df.surface_total < IQMayor)]
sns.boxplot(data = nuevo_df, x = "surface_total")
```

```python
nuevo_df.surface_covered = nuevo_df.surface_covered.fillna(value=0)
IQMenor, IQMayor = limpieza(nuevo_df.surface_covered)
nuevo_df = nuevo_df[( nuevo_df.surface_covered > IQMenor) & (nuevo_df.surface_covered < IQMayor)]
sns.boxplot(data = nuevo_df, x = "surface_covered")
```

```python
nuevo_df.bathrooms = nuevo_df.bathrooms.fillna(value=0)
IQMenor, IQMayor = limpieza(nuevo_df.bathrooms)
nuevo_df = nuevo_df[( nuevo_df.bathrooms > IQMenor) & (nuevo_df.bathrooms < IQMayor)]
sns.boxplot(data = nuevo_df, x = "bathrooms")
```

Para evaluar los distintos modelos usaremos como metrica principal el R2, dicha metrica nos proporciona la exactitud con la cual
el modelo va a responder a la hora calcular. Esto quiere decir en otras palabras que tan bien los resultados esperados son
replicados por el modelo. El maximo valor a alcanzar es 1 (puede ser negativo) lo que nos dice que el modelo esta dando como

1 - ARBOL

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import max_error
from sklearn.metrics import mean_squared_error
tree_model = DecisionTreeRegressor(max_depth=16, random_state=42)
tree_model.fit(X_train, y_train)
y_train_pred_tree = tree_model.predict(X_train)
y_test_pred_tree = tree_model.predict(X_test)
```

Con el siguiente codigo buscamos el numero optimo de profundidad en el arbol

```python
RMSE_train = []
RMSE_test = []

for max_depth in range(1,30):
    clf = DecisionTreeRegressor(max_depth=max_depth)
    model = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    RMSE_train.append(mean_squared_error(y_train, clf.predict(X_train)))
    RMSE_test.append(mean_squared_error(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(range(1,30), RMSE_train, label='RMSE_train')
plt.plot(range(1,30), RMSE_test, label='RMSE_test')
plt.legend(loc="best")
plt.title("RMSE Training vs RMSE Testing para árboles de decisión")
plt.show()
print(RMSE_test.index(min(RMSE_test))) #Posicion 16 es max_depths=16
```

Lo que podemos debemos obvervar en este grafico es el momento en cual los resultados dejan de aumentar conjuntamente

### PIPELINE

El desarrollo siguente lo hago porque me surgio una duda a la hora de predecir datos nuevos, fue a los datos voy a tener que centrarlos y escalarlos manualemnte?, por eso comence a investigar e implemente un pipeline pensado en datos futuros.

```python
def pipe(features):
    
     
    
    "Defino esta funcion con la intencion de que ante la llegada de nuevos datos no sea necesario realizar todo el proceso de escalado y encodeado.Esto me va a facilitar el trabajo a la hora de predecir"
    "Para esto los datos nuevos deben ser ingresador como un dataframe"
    
    
    from sklearn.compose import ColumnTransformer
    from sklearn.pipeline import Pipeline
    from sklearn.impute import SimpleImputer
    from sklearn.preprocessing import StandardScaler, OneHotEncoder
    from sklearn.model_selection import train_test_split
    
    X= features.loc[:,:]
    
    
    columnas = X.columns.tolist()
    lista_numerica_pipe = []
    lista_categorica_pipe = []
    for x in range(0,len(X.columns))  :
        if X[columnas[x]].dtype == 'int64'or X[columnas[x]].dtype == 'float64':
            lista_numerica_pipe.append(columnas[x])
        else:
            lista_categorica_pipe.append(columnas[x])
    
    cnts_pipeline = Pipeline([('scale', StandardScaler())])
    categ_pipeline = Pipeline([('onehot', OneHotEncoder(handle_unknown='ignore'))])
    
    preprocess_pipeline = ColumnTransformer([
    ('continuous', cnts_pipeline, lista_numerica_pipe),
    ('cat', categ_pipeline, lista_categorica_pipe)], remainder = 'passthrough') 
    
    X_processed = preprocess_pipeline.fit_transform(X)

    
    return X_processed
```
